```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
plt.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

#뇌파 데이터 불러오기]
df = pd.read_csv('./good_bad.csv')
```

```python
df
```

```python
bad=df[df.columns.difference(['species'])][100:200]
good=df[df.columns.difference(['species'])][0:100]
```

```python
print(np.array(good).max(), np.array(good).min(), np.array(bad).max(), np.array(bad).min())
```

```python
print(np.array(good).mean(), np.array(bad).mean())
```

```python
fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(111)
x=np.array([i+1 for i in range(100)])
ax1.plot(x, good, 'bo--', label="good", linewidth=2)
ax1.set_xlabel('시간')
ax1.set_ylabel('good')
ax1.set_ylim(-2050, 2050)

ax2 = ax1.twinx()
ax2.plot(x, bad, 'r<-', label="bad", linewidth=2)
ax2.set_ylabel('bad')
ax2.set_ylim(-2050,2050)

plt.show()

#good -> blue, bad -> red
```
